```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
df=pd.read_csv("/content/drive/MyDrive/for-ml-work/dataset/dataset.csv")
encoder_labels=[]
col_names=[]
```

```python
dlist=df.Disorder.to_list()
random.shuffle(dlist)
dlist=dlist[0:7000]
x1=dlist.count('Anxiety')
y1=dlist.count('Normal')
z1=dlist.count('Stress')
a1=dlist.count('Depression')
b1=dlist.count('Loneliness')
l=[x1,y1,z1,a1,b1]
y=np.array(l)
plt.subplot(1,2,1)
plt.pie(y,labels=['Anxiety','Normal', 'Stress', 'Depression', 'Loneliness'], autopct='%1.1f%%',colors=['red','lightblue','darkslategray','olive','yellow'],radius=2)
plt.subplots_adjust(left=0.2,bottom=0.1, right=2,top=0.9,wspace=0.4,hspace=0.4)
plt.subplot(1,2,2)
plt.bar(['Anxiety','Normal','Stress','Depression','Loneliness'],y,color=['red','lightblue','darkslategray','olive','yellow'])
plt.xlabel('Disorder')
plt.ylabel('No. of records')
plt.show()
```
