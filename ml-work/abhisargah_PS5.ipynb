```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
df=pd.read_csv("/content/drive/MyDrive/for-ml-work/dataset/dataset.csv")
encoder_labels=[]
col_names=[]
```

```python
dlist=df.Disorder.to_list()
random.shuffle(dlist)
dlist=dlist[0:7000]
x1=dlist.count('Anxiety')
y1=dlist.count('Normal')
z1=dlist.count('Stress')
a1=dlist.count('Depression')
b1=dlist.count('Loneliness')
l=[x1,y1,z1,a1,b1]
y=np.array(l)
plt.subplot(1,2,1)
plt.pie(y,labels=['Anxiety','Normal', 'Stress', 'Depression', 'Loneliness'], autopct='%1.1f%%',colors=['red','lightblue','darkslategray','olive','yellow'],radius=2)
plt.subplots_adjust(left=0.2,bottom=0.1, right=2,top=0.9,wspace=0.4,hspace=0.4)
plt.subplot(1,2,2)
plt.bar(['Anxiety','Normal','Stress','Depression','Loneliness'],y,color=['red','lightblue','darkslategray','olive','yellow'])
plt.xlabel('Disorder')
plt.ylabel('No. of records')
plt.show()
```

```python
print("\t\t\t\t\t Factors determining Anxiety are:")
x=np.array(df[['feeling.nervous']])
y=np.array(df[['breathing.rapidly']])
z=np.array(df[['having.trouble.in.sleeping']])
a=np.array(df[['panic']])
b=np.array(df[['sweating']])
c=np.array(df[['trouble.in.concentration']])

fig = plt.figure()
ax1 = fig.add_subplot(231)
ax2 = fig.add_subplot(232)
ax3 = fig.add_subplot(233)
ax4 = fig.add_subplot(234)
ax5 = fig.add_subplot(235)
ax6 = fig.add_subplot(236)

ax1.hist(x,color='red')
ax2.hist(y,color='red')
ax3.hist(z,color='red')
ax4.hist(a,color='red')
ax5.hist(b,color='red')
ax6.hist(c,color='red')

# Set common labels
fig.text(1.1, 0.009, 'Yes/No', ha='center', va='center')
fig.text(0.02, 0.5, 'Number of Records', ha='center', va='center', rotation='vertical')

ax1.set_title('Nervous')
ax2.set_title('Rapid Breathing')
ax3.set_title('Troubled Sleeping')
ax4.set_title('Panic')
ax5.set_title('Sweating')
ax6.set_title('Troubled Concentration')

plt.subplots_adjust(left=0.2, bottom=0.1, right=2, top=0.9, wspace=0.4, hspace=0.4)

plt.show()
```
